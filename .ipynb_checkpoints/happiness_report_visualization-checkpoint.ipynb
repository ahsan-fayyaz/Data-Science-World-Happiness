```python
g0 = sb.pairplot(data=CWH, kind='scatter', height = 2, hue = "life_ladder",
            x_vars=['life_ladder'], y_vars=['gdp_per_capita'])

g1 = sb.pairplot(data=CWH, kind='scatter', height = 2, hue = "life_ladder",
            x_vars=['life_ladder'], y_vars=['social_support'])

g0.savefig('g0.png', dpi=300)
plt.close(g0.fig)

g1.savefig('g1.png', dpi=300)
plt.close(g1.fig)


############### 2. CREATE YOUR SUBPLOTS FROM TEMPORAL IMAGES
f, axarr = plt.subplots(1, 2, figsize=(40, 40))

axarr[0].imshow(mpimg.imread('g0.png'))
axarr[1].imshow(mpimg.imread('g1.png'))

# turn off x and y axis
[ax.set_axis_off() for ax in axarr.ravel()]

plt.tight_layout()
plt.show()
```
