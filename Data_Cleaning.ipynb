```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
WHR = pd.read_excel("WHR2021.xls")
```

```python
WHR.info()
```

```python
# Checking for null values
WHR.isnull().sum()
```

```python
#Drop null values
WHR.dropna(inplace = True)
```

```python
# Check if null values still exists
WHR.isnull().sum()
```

```python
WHR_Corr = WHR.corr(method = "pearson")
plt.figure(figsize = (20,10))
sb.heatmap(WHR_Corr, annot = True)
```

```python
WHR_Corr = WHR.corr(method = "kendall")
plt.figure(figsize = (20,10))
sb.heatmap(WHR_Corr, annot = True)
```
